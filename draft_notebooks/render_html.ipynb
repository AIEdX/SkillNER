```python
from typing import List
from IPython.core.display import HTML


def document(children: List = []):
    content = f"""
        <head>
            <link
                id="external-css"
                rel="stylesheet"
                type="text/css"
                href="https://unpkg.com/tailwindcss@^2/dist/tailwind.min.css"
                media="all"
            />
        </head>

        <body>
            {" ".join(children)}
        </body>
    """
    return HTML(content)

def element(ele_type: str ="div", className: str = "", children: List = [], **kwargs):
    other_props = ""

    for key, val in kwargs:
        other_props += f"{key}={val}"

    content = f"""
        <{ele_type} class={className} {other_props}>
            {" ".join(children)}
        </{ele_type}>
    """
    return content

document(children=[
    element(ele_type="span", children=[
        "Badr MOUFAD", 

        element(ele_type="span", children=[
            "Mahmoud MOUFAD"
        ])
    ])
])
```

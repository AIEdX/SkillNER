```python
n_grams = [skills_db[key] for key in skills_db if skills_db[key]['skill_len']>1  ]
```

```python
import re
rx = r"\b[A-Z](?=([&.]?))(?:\1[A-Z])+\b"
s =  '.NET Framework 1'

def extract_sub_forms(skill_name):
    return [x.group() for x in re.finditer(rx, skill_name)] 
def remove_btwn_par(str_):
    return re.sub("[\(\[].*?[\)\]]", "", str_)
extract_sub_forms(s)    
```

```python
c=0
subs=[]
for n_skill in n_grams :
    skill_name = n_skill['skill_name']
    new_skill_name  = remove_btwn_par(skill_name)
    sub_f = extract_sub_forms(new_skill_name)
    if sub_f!=[]:
        print(skill_name)
        print(sub_f)
        print('--------')
        for s in sub_f : 
            subs.append(s)
        c+=1
        
```

```python
import collections
dist = collections.Counter(subs)
```

```python
for key in skills_db :
    if skills_db[key]['skill_len']>2:
        
        skill_name = skills_db[key]['skill_name']
        new_skill_name  = remove_btwn_par(skill_name)
        skill_low = skills_db[key]['low_surface_forms']
        sub_abv = extract_sub_forms(new_skill_name)
        for abv in  sub_abv : 
            if dist[abv]==1:
                skill_low.append(abv)
        skills_db[key]['low_surface_forms'] =  skill_low
                
        
    
    
```
